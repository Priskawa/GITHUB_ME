```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 100

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 4, 5, 3, 4, 5, 7, 2]
y2 = [3, 5, 6, 4, 5, 6, 8, 3]
y3 = [4, 6, 7, 5, 6, 7, 9, 4]
y4 = [5, 7, 8, 6, 7, 8, 10, 5]

ax.plot(x, y, label="Grafik: -", linestyle="-", color=(1, 0.5, 0))
ax.plot(x, y2, label="Grafik: :", linestyle=":")
ax.plot(x, y3, label="Grafik: -.", linestyle="-.")
ax.plot(x, y4, label="Grafik: --", linestyle="--", linewidth=6)

ax.legend()
ax.set_ylim([0, 11])
ax.set_xlim([1, 8])
ax.set_title("Farben und Linienstile")

# Hintergrundfarbe anpassen
ax.set_facecolor((0.95, 0.95, 0.95))

plt.show()
```
