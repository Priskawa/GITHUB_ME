```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 100

x = [2011, 2013, 2015, 2017]
y = [1050, 1400, 1300, 1600]
y2 = [1100, 1300, 1100, 1400]

ax.bar(x, y, 
       label="Gewinn x1000", 
       tick_label=["Jahr 2011", "Jahr 2013", "Jahr 2015", "Jahr 2017"],
       width=-0.4,
       align="edge"
)
ax.bar(x, y2, 
       label="Gewinn 2. Firma x1000", 
       tick_label=["Jahr 2011", "Jahr 2013", "Jahr 2015", "Jahr 2017"],
       width=0.4,
       align="edge"
)

ax.legend()
ax.set_title("Balkendiagramm")

plt.show()
```
