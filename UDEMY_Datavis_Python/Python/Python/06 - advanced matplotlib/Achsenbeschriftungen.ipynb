```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 100

x = [2011, 2013, 2015, 2017]
y = [1050, 1400, 1300, 1600]

ax.plot(x, y, label="Gewinn x1000")

ax.set_xticks([2011, 2013, 2015, 2017])
ax.set_xticklabels(["Jahr 2011", "Jahr 2013", "Jahr 2015", "Jahr 2017"], rotation=45)

ax.legend()
ax.set_title("Punkte mit einzeichnen")

plt.show()
```
