```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 100


ax.pie([100, 50, 33], 
       labels=["Option 1", "Option 2", "Option 3"],
       startangle=45,
       shadow=True,
       autopct="%.1f%%", # autopct="%i%%",
       explode=[0, 0, 0.1],
       colors=[(1, 0, 0), (1, 1, 0), (0, 0, 1)]
)

ax.axis("equal")

ax.set_title("Kreisdiagramm")

plt.show()
```
