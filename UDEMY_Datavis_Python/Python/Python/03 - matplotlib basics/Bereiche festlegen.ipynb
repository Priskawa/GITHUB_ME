```python
%matplotlib inline

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fig.dpi = 100

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 4, 5, 3, 4, 5, 7, 2]

ax.plot(x, y)

ax.set_ylim([0, 8])
ax.set_xlim([1, 8])
ax.set_title("Anzahl Neukunden im Verlauf der Zeit")
ax.set_xlabel("Datum")
ax.set_ylabel("Anzahl an Neukunden")

plt.show()
```
