# Openstreetmap - Daten

```python
import cartopy.crs as ccrs
from cartopy.io.img_tiles import OSM, GoogleTiles
import matplotlib.pyplot as plt


tiles = OSM()

plt.figure(figsize=(16, 16))

ax = plt.axes(projection=tiles.crs)

# Hier kannst du den Ausschnitt definieren
ax.set_extent([5.229, 15.952, 46.438, 56.097])

# Die Zahl entspricht dem Zoomlevel. 
# => Geringes Level = Weniger Daten mssen bertragen werden, ggf. etwas verpixelt
# => Hohes Level => Mehr Daten mssen von Openstreetmap geladen werden, tendenziell weniger verpixelt

ax.add_image(tiles, 6)

plt.show()
```

# Natrlich kannst du Openstreetmap auch mit einer Projektion kombinieren :D 

```python
import cartopy.crs as ccrs
from cartopy.io.img_tiles import OSM, GoogleTiles
import matplotlib.pyplot as plt


tiles = OSM()

plt.figure(figsize=(16, 16))

#ax = plt.axes(projection=tiles.crs)
ax = plt.axes(projection=ccrs.NearsidePerspective(
    central_longitude=-0.1071, 
    central_latitude=51.4899,
    satellite_height=35785831
))

# ax.set_extent([5.229, 15.952, 46.438, 56.097])

ax.add_image(tiles, 4)

plt.show()
```
