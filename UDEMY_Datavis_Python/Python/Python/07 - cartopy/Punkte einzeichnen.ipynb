## Punkte auf Karte einzeichnen

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig = plt.figure(figsize=(10, 15))
ax = plt.axes(projection=ccrs.Robinson())

ax.set_global()
ax.coastlines(resolution="110m")
ax.gridlines()

ax.plot([-0.1071], [51.4899], marker="X", color=(1, 0, 0), transform=ccrs.PlateCarree())

plt.show()
```

## Verbindung zwischen 2 Punkten einzeichnen

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig = plt.figure(figsize=(10, 15))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_global()
ax.coastlines(resolution="110m")
ax.gridlines()

ax.plot([-0.1071], [51.4899], marker="X", color=(1, 0, 0), transform=ccrs.PlateCarree())
ax.plot([114.1689], [22.2935], marker="X", color=(0, 1, 0), transform=ccrs.PlateCarree())

x = [-0.1071, 114.1689]
y = [51.4899, 22.2935]

ax.plot(x, y, color=(0, 0, 1), transform=ccrs.PlateCarree())

plt.show()
```

### Aber ist das jetzt die kürzeste Strecke zwischen den beiden Punkten?

Nein, ist es nicht!

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig = plt.figure(figsize=(10, 15))
ax = plt.axes(projection=ccrs.NearsidePerspective(central_longitude=54.67, central_latitude=33.28))

ax.set_global()
ax.coastlines(resolution="110m")
ax.gridlines()

ax.plot([-0.1071], [51.4899], marker="X", color=(1, 0, 0), transform=ccrs.PlateCarree())
ax.plot([114.1689], [22.2935], marker="X", color=(0, 1, 0), transform=ccrs.PlateCarree())

x = [-0.1071, 114.1689]
y = [51.4899, 22.2935]

ax.plot(x, y, color=(0, 0, 1), transform=ccrs.PlateCarree())
ax.plot(x, y, color=(0, 1, 0), transform=ccrs.Geodetic())


plt.show()
```
