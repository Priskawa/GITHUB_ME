```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

fig = plt.figure(figsize=(5, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

shpfilename = shpreader.natural_earth(resolution="10m", category="cultural", name="admin_0_countries")

reader = shpreader.Reader(shpfilename)

#for geometry in reader.geometries():
#    ax.add_geometries(geometry, crs=ccrs.PlateCarree(), facecolor=(0.9, 0.9, 0.9), edgecolor=(0.7, 0.7, 0.7))

for country in reader.records():
    name = ""
    if "NAME" in country.attributes:
        name = country.attributes["NAME"]
    elif "name" in country.attributes:
        name = country.attributes["name"]
    
    if name == "Germany":
        ax.add_geometries(
            country.geometry, 
            crs=ccrs.PlateCarree(), 
            facecolor=(0.25, 0.25, 0.25), 
            edgecolor=(0.7, 0.7, 0.7)
        )
    else:
        ax.add_geometries(
            country.geometry, 
            crs=ccrs.PlateCarree(), 
            facecolor=(0.9, 0.9, 0.9), 
            edgecolor=(0.7, 0.7, 0.7)
        )
    

# http://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/cultural/ne_10m_admin_0_countries.zip

ax.set_extent([5.229, 15.952, 46.438, 56.097])
# ax.coastlines(resolution="110m")
ax.gridlines()
ax.set_aspect(1.33)

plt.show()
```
