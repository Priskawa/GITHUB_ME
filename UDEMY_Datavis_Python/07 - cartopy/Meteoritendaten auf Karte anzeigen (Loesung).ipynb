# Aufgabe: Meteoritendaten auf Karte einzeichnen

Schau dir nochmal die Meteoritendaten an (meteorite-landings.csv). Lese diese Datei ein, und zeichne alle Meteoriten auf die Weltkarte ein.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

fig = plt.figure(figsize=(15, 20))
ax = plt.axes(projection=ccrs.Robinson())

ax.set_global()
ax.coastlines(resolution="50m")

ax.add_feature(cfeature.OCEAN, zorder=0)
ax.add_feature(cfeature.LAND, zorder=0)
ax.add_feature(cfeature.BORDERS, zorder=0, edgecolor='gray')

ax.gridlines()

plt.show()
```

```python
import pandas as pd

df = pd.read_csv("./../data/meteorite-landings.csv")
df.head()
```

```python
df2 = df[df["reclat"].notnull()]
df3 = df2[df2["reclong"].notnull()]

long = df3["reclong"]
lat = df3["reclat"]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

fig = plt.figure(figsize=(15, 20))
ax = plt.axes(projection=ccrs.Robinson())

ax.set_global()
ax.coastlines(resolution="50m")


ax.plot(long, lat, linestyle=" ", marker="X", transform=ccrs.PlateCarree(), color=(0, 0, 0), alpha=0.2)

ax.add_feature(cfeature.OCEAN, zorder=0)
ax.add_feature(cfeature.LAND, zorder=0)
ax.add_feature(cfeature.BORDERS, zorder=0, edgecolor='gray')

ax.gridlines()

plt.show()
```
