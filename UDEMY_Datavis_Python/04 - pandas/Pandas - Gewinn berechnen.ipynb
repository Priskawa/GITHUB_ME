```python
import pandas as pd
```

```python
df = pd.read_excel("./Firmeneinnahmen.xlsx")
```

```python
df.head()













```

```python
df2 = df[df["Jahr"] != "Gesamt:"]
df3 = df2[df2["Jahr"].notnull()]
```

```python
gewinn = df3["Einnahmen"] - df3["Ausgaben"]
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 80

ax.plot(df3["Jahr"], df3["Einnahmen"])
ax.plot(df3["Jahr"], df3["Ausgaben"])
ax.plot(df3["Jahr"], gewinn, label="Gewinn")

ax.legend()

plt.show()
```
