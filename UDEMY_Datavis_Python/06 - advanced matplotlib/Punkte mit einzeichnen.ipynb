```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 100

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 4, 5, 3, 4, 5, 7, 2]


ax.plot(x, y, label="Grafik", marker="o", markerfacecolor="red", markersize=8)

ax.legend()
ax.set_title("Punkte mit einzeichnen")

plt.show()
```
