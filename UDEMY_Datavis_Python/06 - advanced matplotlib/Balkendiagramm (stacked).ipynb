# Balkendiagramm: Stacked Barplot

Mit einem gestapelten Diagramm kannst du super z.B. einen Anteil einer Gesch√§ftssparte am Gesamtgewinn der Firma visualisieren. Machen wir mal ein Beispiel...


```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

fig.dpi = 100

x = [2011, 2013, 2015, 2017]
y = [1050, 1400, 1300, 1600] # Gewinn der Firma, exklusive Gesch√§ftszweig X
y2 = [200, 100, 200, 150]    # Gewinn der Firma durch Gesch√§ftszweig X

ax.bar(x, y, 
       label="Gewinn x1000", 
       tick_label=["Jahr 2011", "Jahr 2013", "Jahr 2015", "Jahr 2017"]
)
ax.bar(x, y2, bottom=y)

ax.legend()
ax.set_title("Balkendiagramm")

plt.show()
```
