## Charts aufhübschen mit Seaborn

Die Library Seaborn bietet viele Möglichkeiten, Charts besser darzustellen (statistische Datenvisualisierung)

**import seaborn as sns** - Lib laden

```python
import seaborn as sns
```

```python
import seaborn as sns

%matplotlib inline
import matplotlib.pyplot as plt


sns.set_style("darkgrid") # white, whitegrid, dark, darkgrid

fig, ax = plt.subplots()

fig.dpi = 100

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 4, 5, 3, 4, 5, 7, 2]
y2 = [4, 10, 7, 10, 4, 6, 10, 4]

ax.plot(x, y, label="Neukunden")
ax.plot(x, y2, label="Interessenten")

ax.legend()
ax.set_ylim([0, 11])
ax.set_xlim([1, 8])
ax.set_title("Performance im Laufe der Zeit")
ax.set_xlabel("Datum")
ax.set_ylabel("Anzahl")

plt.show()
```
