```python
import json
import urllib.request

your_url = 'https://app-prod-static-voteinfo.s3.eu-central-1.amazonaws.com/v1/ogd/sd-t-17-02-20231126-kantAbstimmung.json'
data = ""
with urllib.request.urlopen(your_url) as url:
    data = json.loads(url.read().decode())
```

```python
kantone = data['kantone']
#kantone = vorlagen[0]['kantone']
print(kantone)
```

```python
gemres=[]
attributes = ["gebietAusgezaehlt",
"jaStimmenInProzent",
"jaStimmenAbsolut",
"neinStimmenAbsolut",
"stimmbeteiligungInProzent"]
#spaltennamen=["Gemeindename","Gemeindenummer"]+attributes+["neinStimmenInProzent"]
spaltennamen=["Gemeindename","Gemeindenummer","Ausgezählt","Ja Prozent","Nein Prozent","Ja-Stimmen","Nein-Stimmen","Stimmbeteiligung"]
print(spaltennamen)
gemres.append(spaltennamen)
```

```python
for kanton in kantone:

  kantonname = kanton['geoLevelname']
  kantonnr = kanton['geoLevelnummer']

  if kantonnr==3:
    vorlage=kanton['vorlagen'][0]
    gemeinden= kanton['vorlagen'][0]['gemeinden']
    for gemeinde in gemeinden:
      zeile = []
      gemeindename = gemeinde['geoLevelname']
      gemeindenr = gemeinde['geoLevelnummer']
      gemeinderesultat=gemeinde['resultat']
      neinStimmenInProzent=100-gemeinderesultat['jaStimmenInProzent']
      zeile.append(gemeindename)
      zeile.append(gemeindenr)

      for attribut in attributes:
       zeile.append(gemeinderesultat[attribut])

      #zeile.append(neinStimmenInProzent)
      zeile.insert(4,neinStimmenInProzent)



      gemres.append(zeile)

print(gemres)
```

```python
import pandas as pd
import gspread
```

```python
df = pd.DataFrame(gemres)
df.columns = df.iloc[0]
df = df[1:]
print(df)
```
